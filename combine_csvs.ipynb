```python
import pandas as pd
import numpy as np
import os
import glob
import csv
import sys
```

```python
files = [f for f in os.listdir('.') if os.path.isfile(f)]
print(files)
```

```python
merged = []

for f in files:
    filename, ext = os.path.splitext(f)
    if ext == '.csv':
        read = pd.read_csv(f)
        merged.append(read)

result = pd.concat(merged)
result

result.to_csv('merged.csv')
```
