```python
import pandas as pd
import numpy as np
```

```python
#all_logs.csv generated at the command line.
all_logs = pd.read_csv('all_logs.csv')
```

```python
#create dataframe for data manipulation
all_logs_df = pd.DataFrame(all_logs)
```

```python
all_logs_df
```

```python
#drop dummy connections
all_logs_df = all_logs_df[all_logs_df.Host != '127.0.0.1']
```

```python
type(all_logs_df)
```

```python
#drop weird columns generated by cat *.csv > all_logs.csv
all_logs_df = all_logs_df.drop(['Unnamed: 8','Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13','Unnamed: 14','Unnamed: 15','Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20','Unnamed: 21','Unnamed: 22','Unnamed: 23','Unnamed: 24','Unnamed: 25'], axis=1)
```

```python
# count methods used
methods = all_logs_df['Method'].value_counts()
print(methods)
```

```python
# count response codes
response_codes = all_logs_df['Response Code'].value_counts()
print(response_codes)
# this looks weird, but we're primarily interested in 200, 403, and 404
```
